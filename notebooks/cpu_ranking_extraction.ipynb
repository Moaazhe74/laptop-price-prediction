```python
import pandas as pd
base_url = "https://technical.city/en/cpu/rating?pg={}"

all_data = []
page = 1

while True:
    try:
        print(f"Reading page {page}...")
        url = base_url.format(page)

        tables = pd.read_html(url)
        dfy = tables[0]

        # Extract only needed columns
        dfy = dfy[["CPU", "Performance"]]

        # Stop if no data
        if dfy.empty:
            break

        all_data.append(dfy)
        page += 1

    except:
        break

# Combine all pages
final_df_Cpu = pd.concat(all_data, ignore_index=True)

# Remove duplicates
final_df_Cpu = final_df_Cpu.drop_duplicates()

# Rename columns
final_df_Cpu.columns = ["CPU Name", "Performance"]

# Save to Excel
final_df_Cpu.to_excel("cpu_ranking.xlsx", index=False)

print(f"Done! Total CPUs: {len(final_df_Cpu)}")
```
