```python
import pandas as pd
base_url = "https://technical.city/en/video/rating?pg={}"

all_data = []

page = 1

while True:
    try:
        print(f"Reading page {page}...")
        url = base_url.format(page)

        tables = pd.read_html(url)
        df_Gpu = tables[0]

        # Keep only needed columns
        df_Gpu = df_Gpu[["Graphics card", "Performance"]]

        # Stop if empty or repeated
        if df_Gpu.empty:
            break

        all_data.append(df_Gpu)

        page += 1

    except:
        break

# Combine all pages
final_df_Gpu = pd.concat(all_data, ignore_index=True)

# Remove duplicates just in case
final_df_Gpu = final_df_Gpu.drop_duplicates()

# Rename columns
final_df_Gpu.columns = ["GPU Name", "Performance"]

# Save to Excel
final_df_Gpu.to_excel("all_gpu_ranking_1566.xlsx", index=False)

print(f"Done! Total GPUs saved: {len(final_df_Gpu)}")
final_df_Gpu.head()
```
